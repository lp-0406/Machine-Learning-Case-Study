**data preprocessing- standardization**

>we scale the features so they have a mean of 0 and standard deviation of 1.

WHY??


1.   The feature values represent pixel intensity values across 36 bands.

2. Some of these bands (like band_7, band_36) can have much higher ranges than others ie one may have from 10-1000 while other bands have from 50-5000.

3. Models like Logistic Regression and AdaBoost are sensitive to magnitude differences, which can make features with larger values dominate learning

**so we use standardization to:**



*   Equalize all feature scales
*  Allow fair contribution from each pixel band to the model

* Improve models like Logistic Regression









```python
#standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Outliers**


*   In classification tasks (especially in satellite image data), outliers might actually be valid class boundaries.

eg: a pixel with extreme intensity may indicate a different class (soil/vegetation) rather than an error.

**so outliers arent handles to avoid loss of data**

**PCA (Principal Component Analysis)**
This dataset has 36 features, which is not extremely high-dimensional so no pca is done

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import classification_report
```

**80/20 split**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

**Logistic Regression**


>Itâ€™s a baseline classifier.
>Compare simple linear models vs. complex ones

* dataset has non-linearly separable pixel classes
* logistic regression may struggle because : the relationship between pixel values and classes is complex

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_acc = accuracy_score(y_test, lr_pred)

print(f"Logistic Regression Accuracy: {lr_acc:.4f}")
print("\nClassification Report:\n")
print(classification_report(y_test, lr_pred))
```

**Decision Trees**
>Works well when there are non-linear relationships,ie the pixel intensities of satellite images

* Can capture complex patterns across bands without needing to scale the data

* Provides interpretability â€” we can see how the decision is made from bands

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)
print(f"Accuracy: {dt_acc:.4f}")
print("Classification Report:")
print(classification_report(y_test, dt_pred))
```

**Random Forest Classifier**
> * builds many decision trees using random feature subsets
* Reduce overfitting from a single decision tree
* Improve accuracy by combining decisions


**used in landsat cuz**
* The pixel patterns can vary a lot between classes

* Some bands might be noisy, but Random Forest ignores those through randomness

* More stable and less variance-sensitive, which suits high-dimensional image data

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
print(f"Accuracy: {rf_acc:.4f}")
print("Classification Report:")
print(classification_report(y_test, rf_pred))
```

**AdaBoost**
>  focuses on misclassified points and improves them iteratively

* Helpful when thereâ€™s class overlap or edge cases, like mixed pixels in satellite rows

```python
ab = AdaBoostClassifier(n_estimators=100, random_state=42)
ab.fit(X_train, y_train)
ab_pred = ab.predict(X_test)
ab_acc = accuracy_score(y_test, ab_pred)
print(f"Accuracy: {ab_acc:.4f}")
print("Classification Report:")
print(classification_report(y_test, ab_pred))
```

```python
print(f"Logistic Regression Accuracy: {lr_acc:.4f}")
print(f"Decision Tree Accuracy: {dt_acc:.4f}")
print(f"Random Forest Accuracy: {rf_acc:.4f}")
print(f"AdaBoost Accuracy: {ab_acc:.4f}")
```

**60/40 split**

**Reason 1: Landsat has moderately sized data (6435 samples)**

* Using a larger test set (40%) gives us a more reliable estimate of how well the model generalizes to unseen satellite data.

* Satellite data often contains variability and noise in spectral bands, so itâ€™s important to stress-test the model on more test data.

**Reason 2:** **Less training data = testing model robustness**
* Training on 60% tests the model's ability to learn patterns efficiently from fewer samples.

* For satellite classification, it helps check whether the model can still detect patterns across bands with limited pixel training data.

```python
X_train_60, X_test_40, y_train_60, y_test_40 = train_test_split(X_scaled, y, test_size=0.4, random_state=42)
```

```python
lr_60 = LogisticRegression(max_iter=1000)
lr_60.fit(X_train_60, y_train_60)
lr_pred_60 = lr_60.predict(X_test_40)
lr_acc_60 = accuracy_score(y_test_40, lr_pred_60)
```

```python
dt_60 = DecisionTreeClassifier(random_state=42)
dt_60.fit(X_train_60, y_train_60)
dt_pred_60 = dt_60.predict(X_test_40)
dt_acc_60 = accuracy_score(y_test_40, dt_pred_60)
```

```python
rf_60 = RandomForestClassifier(n_estimators=100, random_state=42)
rf_60.fit(X_train_60, y_train_60)
rf_pred_60 = rf_60.predict(X_test_40)
rf_acc_60 = accuracy_score(y_test_40, rf_pred_60)
```

```python
ab_60 = AdaBoostClassifier(n_estimators=100, random_state=42)
ab_60.fit(X_train_60, y_train_60)
ab_pred_60 = ab_60.predict(X_test_40)
ab_acc_60 = accuracy_score(y_test_40, ab_pred_60)
```

```python
print("60/40 Split Accuracies:")
print(f"Logistic Regression Accuracy: {lr_acc_60:.4f}")
print(f"Decision Tree Accuracy:{dt_acc_60:.4f}")
print(f"Random Forest Accuracy:{rf_acc_60:.4f}")
print(f"AdaBoost Accuracy:{ab_acc_60:.4f}")
```

```python
X_train_75, X_test_25, y_train_75, y_test_25 = train_test_split(X_scaled, y, test_size=0.25, random_state=42)

lr_75 = LogisticRegression(max_iter=1000)
lr_75.fit(X_train_75, y_train_75)
lr_pred_75 = lr_75.predict(X_test_25)
lr_acc_75 = accuracy_score(y_test_25, lr_pred_75)

dt_75 = DecisionTreeClassifier(random_state=42)
dt_75.fit(X_train_75, y_train_75)
dt_pred_75 = dt_75.predict(X_test_25)
dt_acc_75 = accuracy_score(y_test_25, dt_pred_75)

rf_75 = RandomForestClassifier(n_estimators=100, random_state=42)
rf_75.fit(X_train_75, y_train_75)
rf_pred_75 = rf_75.predict(X_test_25)
rf_acc_75 = accuracy_score(y_test_25, rf_pred_75)

ab_75 = AdaBoostClassifier(n_estimators=100, random_state=42)
ab_75.fit(X_train_75, y_train_75)
ab_pred_75 = ab_75.predict(X_test_25)
ab_acc_75 = accuracy_score(y_test_25, ab_pred_75)

print("75/25 Split Accuracies:")
print("Logistic Regression Accuracy:", lr_acc_75)
print("Decision Tree Accuracy:", dt_acc_75)
print("Random Forest Accuracy:", rf_acc_75)
print("AdaBoost Accuracy:", ab_acc_75)
```

```python
accuracies = {
    "Logistic Regression": [lr_acc_60, lr_acc, lr_acc_75],
    "Decision Tree": [dt_acc_60, dt_acc, dt_acc_75],
    "Random Forest": [rf_acc_60, rf_acc, rf_acc_75],
    "AdaBoost": [ab_acc_60, ab_acc, ab_acc_75]
}
```

```python
splits = ["60/40", "72/25", "75/25"]

best_model = None
best_accuracy = 0.0
best_split = None

for model, scores in accuracies.items():
    for i, score in enumerate(scores):
        if score > best_accuracy:
            best_accuracy = score
            best_model = model
            best_split = splits[i]

print(f"\nHighest Accuracy: {best_model} with {best_accuracy} on {best_split} split")
```
